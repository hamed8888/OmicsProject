```python
import pandas as pd
import json
file = 'sleep-2022-11-22.json'
with open(file) as sleepData:
    jsonVal = json.load(sleepData)

df = pd.json_normalize(jsonVal)
# print(df.iloc[1])
df
```

```python
df['dateOfSleep']= pd.to_datetime(df['dateOfSleep'])
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import Range1d
import panel as pn
pn.extension()
# output to static HTML file
#output_file("line.html")
def drawPlot(column):
    p = figure(sizing_mode="stretch_width",toolbar_location="below" ,x_axis_type="datetime")
    p.y_range = Range1d(0, 1000)
    # add a circle renderer with a size, color, and alpha
    p.circle(df["dateOfSleep"], df[column],legend_label=column, size=8, color="green", alpha=0.5)
    p.line(df["dateOfSleep"], df["timeInBed"],legend_label="timeInBed", line_width=2, color="blue", alpha=0.5)
    p.circle(df["dateOfSleep"], df["timeInBed"],legend_label="timeInBed", size=8, color="blue", alpha=0.5)

    p.legend.location = "top_right"
    # show the results
    return p
```

```python
select = pn.widgets.Select(name='Select', options=["duration", "minutesAsleep","levels.summary.rem.minutes"])
```

```python
pn.extension()
pn.interact(drawPlot, column= select)
```
